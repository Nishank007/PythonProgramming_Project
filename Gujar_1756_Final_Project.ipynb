#### -> Using lxml.html parser, we webscrape the CNN Money Market Movers Website and stored the data in a dataframe.
#### -> We divide the dataframe into Most Active, Gainers and Losers ,and Print them to the user.
#### -> The user enters a specific stock they are interested in. We then webscrape the page with that particular stock using BeautifulSoup.
#### -> We then collect information such as OPEN price, PREV CLOSE price, VOLUME and MARKET CAP from the stocks webpage.
#### -> This data is then displayed to the user and saved in stocks.csv file.

```python
import requests
import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup
import csv

url='https://money.cnn.com/data/hotstocks/' ## Url for CNN money market movers wesite
page = requests.get(url) 
doc = lh.fromstring(page.content) ## To get data stored between <tr></tr> on the webpage
tr_elements = doc.xpath('//tr')

tr_elements = doc.xpath('//tr')
col=[]
i=0 
for t in tr_elements[0]: ## For each row, store header and an empty list
    i+=1
    name=t.text_content()
    col.append((name,[]))
    
for j in range(1,len(tr_elements)): ## Since the first row is the header, data is stored from second row onwards
    T=tr_elements[j]
    if len(T)!=4: ## To make sure we do not add any unnecessary data to our dataframe
        break
    i=0
    for t in T.iterchildren():
        data=t.text_content() 
        if i>0:
            try:
                data=int(data)
            except:
                pass
        col[i][1].append(data)
        i+=1

Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict) ## Our table is stored in a dataframe df

ma = df.loc[:9] ## Dividing our dataframe in most active, gainers & losers
tg = df.loc[11:20]
tl = df.loc[22:]

print("This is a program to scrape data from the 'https://money.cnn.com/data/hotstocks/'  for a class project. \n")
print("Which stock are you interested in: \n")

print(" Most Actives: ")
for m in ma['Company']:
    print(m)
print("\n Gainers: ")
for g in tg['Company']:
    print(g)
print("\n Losers: ")
for l in tl['Company']:
    print(l)

user = input('\n User inputs: ') ## Asking the user for stocks he/she wants to view

url1 = "https://money.cnn.com/quote/quote.html?symb="+user ## Url for the individual stock
page1 = requests.get(url1)
soup = BeautifulSoup(page1.text, 'lxml')
stock_value = soup.find('h1', class_ = 'wsod_fLeft').text ## To get data stored between <h1></h1> on the webpage
stock_key = user
trading_info = soup.find('table', class_ = 'wsod_dataTable wsod_dataTableBig')
info_with_label = trading_info.find_all('td')
info_without_label = trading_info.find_all('td', class_='wsod_quoteDataPoint')
full = stock_key+" "+stock_value
print("\nThe data for "+full+" is the following\n")
print(full)
stocks = [stock_key, stock_value] ## Print the data fetched earlier for an individual stock
for i in info_with_label:
    print(i.text)
for i in info_without_label:
    stocks.append(i.text)
    
with open('stocks.csv', 'a+', encoding='UTF8', newline='') as f: ## Saving the stock in stocks.csv
    writer = csv.writer(f)
    writer.writerow(stocks)
```
